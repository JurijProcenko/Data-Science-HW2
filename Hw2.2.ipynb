Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python

```python
python_data = df[df["Мова програмування"]=="Python"]
```

```python
python_data.shape
```

Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"

```python
group_data = python_data.groupby(["Посада"])
```

Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

```python
result_df = python_data.groupby('Посада')['Зарплата на місяць'].agg(['min', 'max']).reset_index()
result_df
```

Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
def fill_avg_salary(row):
    return python_data[python_data['Посада'] == row['Посада']]['Зарплата на місяць'].mean()

python_data["avg"]=python_data.apply(fill_avg_salary, axis=1)
python_data
```

Створіть описову статистику за допомогою методу describe для нового стовпчика.

```python
python_data.avg.describe()
```

Стовпчиковий графік мінімальних зарплат за посадою

```python
plt.figure(figsize=(15, 5))
plt.bar(result_df["Посада"], result_df["min"],color=["b", "r", "y", "g", "c"],)
plt.xlabel("Посада", fontsize=12, color="midnightblue")
plt.ylabel("Гроші", fontsize=12, color="midnightblue")
plt.title("Мінімальні зарплати за посадою", fontsize=15)
plt.legend()
plt.show()
```

Графіки мінімальної та максимальної зарплат за посадою

```python
fig, axs = plt.subplots(2,1, figsize=(15, 8))
axs[0].plot(result_df["Посада"], result_df["min"], marker="v",color="#061358",linestyle="--",)
axs[1].plot(result_df["Посада"], result_df["max"], marker="^",color="#FF5733",)
plt.xlabel("Посада", fontsize=12, color="midnightblue")
plt.ylabel("Гроші", fontsize=12, color="midnightblue")

axs[0].set_title('Мінімальні зарплати за посадою', fontsize=10)
axs[1].set_title('Максимальні зарплати за посадою', fontsize=10)

plt.show()
```
