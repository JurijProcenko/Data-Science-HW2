Прочитайте дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)"

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
source = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C'
df = pd.read_html(source, match='Коефіцієнт народжуваності в регіонах України', thousands=".", decimal=",")[0]
# df = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8")
```

Замініть у таблиці значення "—" на значення NaN

```python
df.replace("—", np.nan, inplace=True)
df
```

Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
df.dtypes
```

Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
df['2014'] = df['2014'].astype('float')
df['2019'] = df['2019'].astype('float')
df.dtypes
```

Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
# Частка у відсотках виглядає краще
df.isnull().sum() / len(df)*100
```

Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
df.drop([len(df)-1], inplace=True)
df
```

Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
for column in df.columns:
    if column == "Регіон": continue

    means = df[column].mean()
    df.fillna({column: means}, inplace=True)
df
```

Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
mean_2019 = df["2019"].mean()
regions = df[df["2019"]>mean_2019]
print(regions["Регіон"])
```

У якому регіоні була найвища народжуваність у 2014 році?

```python
df[df["2014"]==df["2014"].max()]["Регіон"]
```

Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році


```python
plt.barh(df["Регіон"], df["2019"])
plt.xlabel("Країна", fontsize="small", color="midnightblue")
plt.show()
```

Кругова діаграма

```python
plt.pie(df["2019"], labels=df["Регіон"],autopct="%.2f%%",radius=2, explode=[0.05*i for i in range(0, len(df))])
plt.show()
```

Лінійний графік народжуваності по рокам

```python
labels = ["1950", "1960","1970","1990","2000","2012","2014","2019"]
all_together = list(df.sum())
plt.title("Динаміка народжуваності в Україні по рокам", fontsize=15)
plt.plot(labels, all_together[1:])
plt.show()
```
